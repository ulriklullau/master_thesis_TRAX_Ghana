# Code for visualiation of daily and monthly refrigeration loads

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## Necessary data for the plots

```python
# Necessary data for the plots
# Set up time range
hours = [f"{h:02}:00" for h in range(1, 24)] + ["00:00"]

# Set global figure size and resolution for all plots
FIGSIZE = (6, 4)  # inches, good for A4 column width
DPI = 300         # High resolution for thesis figure
```

## Plot 1: Transmission Load Comparison

```python
# 1. Transmission load data (April and August)
transmission_april = [
    0.3148, 0.306, 0.3016, 0.2972, 0.2939, 0.2928, 0.2994, 0.3191,
    0.3378, 0.3588, 0.3776, 0.3941, 0.4063, 0.4129, 0.4162, 0.4118,
    0.4019, 0.3855, 0.3712, 0.3601, 0.3502, 0.3402, 0.3303, 0.3204
]

transmission_august = [
    0.2563, 0.2497, 0.2475, 0.2442, 0.2431, 0.242, 0.2486, 0.2616,
    0.276, 0.2881, 0.3014, 0.3091, 0.3157, 0.3191, 0.3202, 0.3191,
    0.3135, 0.3027, 0.2939, 0.2884, 0.2806, 0.274, 0.2696, 0.2619
]

fig, ax = plt.subplots(figsize=FIGSIZE, dpi=DPI)

ax.plot(hours, transmission_april, label='April', color='orange', linewidth=1.5)
ax.plot(hours, transmission_august, label='August', color='steelblue', linewidth=1.5)

ax.set_title('Hourly Transmission Load (April vs August)')
ax.set_xlabel('Hour of Day')
ax.set_ylabel('Transmission Load [kW]')
ax.grid(True, linestyle='--', alpha=0.6)
ax.set_xticks(hours[::2])  # Show every second hour
ax.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("transmission_load_comparison.png", bbox_inches='tight')
plt.show()
```
